```python
import json
cs=set()
with open('DataCrawling/championInfo.json', 'r') as f:
    d = json.load(f)
    for champ, champ_dict in d['data'].items():
        cs.add(champ_dict['key'])
```

```python
from collections import defaultdict
import surprise as sp
import pandas as pd
import re

ul=[]
cl=[]
ml=[]
ucp=defaultdict(int)
for line in open("DataCrawling/playersInfo.txt"):
    data = line.strip().split('-')
    if len(data) != 3:
        summoner = data[-3].split(',')[-1]
        champ_list = data[-2]
        rank = data[-1]
    else:
        summoner = data[0]
        champ_list = data[1]
        rank = data[2]
    champ_list=champ_list.split(')')
    champ_list=[i.replace('(','').lstrip(',') for i in champ_list if i!='']
    if summoner=="81209212": print(champ_list)
    for i in champ_list:
        if i=='': continue
        p=i.split(',')
        cs.add(p[0])
        if int(p[1])>=7:
            ul.append(summoner)
            cl.append(p[0])
            ml.append(2)
            ucp[str(summoner)+'-'+str(p[0])]=2
        elif int(p[1])>=4:
            ul.append(summoner)
            cl.append(p[0])
            ml.append(1)
            ucp[str(summoner)+'-'+str(p[0])]=1
    for i in cs:
        if ucp[str(summoner)+'-'str(i)]==0:
            ul.append(summoner)
            cl.append(i)
            ml.append(0)
        
            
```

```python
import random
import numpy as np
#t=random.sample(range(0,1000000),200000)
#v=random.sample(range(0,1000000),200000)
ul=np.array(ul)
cl=np.array(cl)
ml=np.array(ml)
mainDict={'ul':ul[:100000],'cl':cl[:100000],'ml':ml[:100000]}
pm=pd.DataFrame(mainDict)
reader=sp.Reader(rating_scale=(0,2))
data=sp.Dataset.load_from_df(pm[['ul', 'cl', 'ml']],reader)
sim_options = {'name': 'cosine',
                'user_based': True 
                }
alg = sp.KNNBasic(sim_options=sim_options)
output=alg.fit(data.build_full_trainset())

t=0
f=0
mse=0
for i in cs:
    p=alg.predict(ul[50000],i)
    r=ucp[str(ul[100001])+'-'str(i)]
    print(p)
    print(r)
    if abs(p.est-r)<0.2:
        t+=1
    else:
        f+=1
    mse+=(p.est-r)**2
print("Correct:"+str(t))
print("Incorrect:"+str(f))
print("Accuracy:"+str(t/(t+f)))
print("MSE:"+str(mse/(t+f)))
```
