```python
def requestMatchData(matchID):
    URL = "https://na1.api.riotgames.com/lol/match/v3/matches/" + matchID + "?api_key=RGAPI-82447fb0-b14e-4e22-9e49-aeaa98ad2afd"
    response = requests.get(URL)
    return response.json()
```

```python
playerSet = set()
```

```python
for i in range(2919197500,2919197600):
    matchID = str(i)
    responseJSON = requestMatchData(matchID)
    data = responseJSON
    if 'participantIdentities' in data:
        print "curr matchID is: ", i
        Identities = data['participantIdentities']
        #playersName = list()
        playersSummonerID = list()
        for i in range (len(Identities)):
            if 'summonerId' not in Identities[i]['player']:
                continue
            playerSet.add(Identities[i]['player']['summonerId'])    

print playerSet
```

```python
f = open("userDataSet1.txt", 'w') # the file to store the output
for p in playerSet:
    f.write(str(p) + "\n")
f.close()
print "done"
```
